```
import matplotlib.pyplot as plt
import numpy as np

x  = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
z= np.random.standard_normal(1000)

plt.hist(z,bins = 20,label="histogram", color="blue")

plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("histogram Example")
plt.legend()
plt.show()
```

```
import matplotlib.pyplot as plt
import numpy as np

# Data for plotting
t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.tan(2 * np.pi * t)

plt.plot(t,s)
plt.grid()
plt.show()

```

```
import matplotlib.pyplot as plt

x1  = [1, 2, 3, 4, 5]
y1 =  [1, 2, 3, 4, 5]

x2  = [1, 2, 3, 4, 5]
y2 = [1, 4, 9, 16, 25]

plt.subplot(1,2,1)
plt.subplots_adjust(wspace=0.5)
plt.plot(x1,y1,'r')
plt.grid()
plt.xlabel('x1 axis')
plt.ylabel('y1 axis')
plt.title('linear')

plt.subplot(1,2,2)
plt.plot(x2,y2,'b')
plt.grid()
plt.xlabel('x2 axis')
plt.ylabel('y2 axis')
plt.title('quadratic')

plt.show()
```
